### EXERCICI 2

Selecciono com a atribut el ranking a la seva lliga (LgRk). Per tal de tenir 2 grups, categoritzo la variable

```python
# Categorització del rankint en 2 grups
def Ranking(x):
    if x<=10:
        return 1;
    elif x>10 and x<=20:
        return 2;
lliga['LgRk_cat']=lliga['LgRk'].map(Ranking)
```

Tenim 2 atributs:
- Categoria 1: a la seva lliga estan classificats a entre els 10 primers
- Categoria 2: a la seva lliga estan classificats a paritr del lloc 11

H0= els equips en la categoria 1 tenen la mateixa mitjana de partits guanyats que els equips de la categoria 2

Ha= els equips de la categoria 1 de mitjana guanyen menys partits que els de la categoria 2

```python
Ranking1=lliga[lliga['LgRk_cat']==1]
Ranking2=lliga[lliga['LgRk_cat']==2]
```

```python
# Primer mirem la normalitat de les 2 categories
stat,p = shapiro(Ranking1['W'])
stat, p
```

```python
stat,p = shapiro(Ranking2['W'])
stat, p
```

```python
Assumim la normalitat de les dues categories
```

```python
# Realitzem el test de igualtat de variances. Utilitzem  Bartlett perquè sabem que les dues categories són normals. 
# Si no fossin normals, podriem realtizar els tests Levene o el test de Fligner-Killeen (són més robustos davant la manca de 
#normalitat)
bartlett_test = stats.bartlett(Ranking1['W'], Ranking2['W'])
bartlett_test
```

```python
# El resultat indica que hem de rebutjar la hipòtesis nul·la d'igualtat de variances. Per tant, no hem d'utilitzar la prova
# t-student i buscar la de t-Welch perquè és robusta davant aquest fet.
```

```python
stats.ttest_ind(Ranking1['W'], Ranking2['W'], alternative='greater',equal_var=False)
```

Rebutjem la hipotèsis nul·la i acceptem que els equips de la categoria 1 guanyen menys partits que els de la categoria 2

### EXERCICI 3

Selecciono com a atribut el número de gols marcats pel jugador marcador top (Top Score players). Per això separo el número de gols marcats del nom del jugador i la categoritzo.

```python
pattern= "([\d]{1,2})"
top=lliga['Top Team Scorer'].str.extract(pattern)
lliga['Top Score players']=top[0]

lliga['Top Score players']=pd.to_numeric(lliga['Top Score players'], errors='coerce')

lliga.isnull().sum()
```

```python
#Per categoritzar
def Top(x):
    if x<=10:
        return 1;
    elif x>10 and x<=20:
        return 2;
    elif x>20:
        return 3;
lliga['Top_cat']=lliga['Top Score players'].map(Top)
```

```python
Top_cat1=lliga[lliga['Top_cat']==1]
Top_cat2=lliga[lliga['Top_cat']==2]
Top_cat3=lliga[lliga['Top_cat']==3]
```

Hi ha 3 atributs:
- Categoria 1: els màxims marcadors marquen 10 gols o menys
- Categoria 2: els màxims marcadors marquen entre 11 i 20 gols
- Categoria 3: els màxims marcadors marquen més de 20 gols

H0: els equips de les 3 categories tenen la mateixa mitjana de gols en contra (GA)
Ha : els equips de les 3 categòries tenen diferent mitjana de gols en contra

```python
# Primer mirem la normalitat de les 3 categories
data=[Top_cat1['GA'],Top_cat2['GA'], Top_cat3['GA']]
for x in data:
    stat, p = shapiro(x)
    print('stat=%.3f, p=%.3f' % (stat, p))
```

Assumim la normalitat de les 3 categories

```python
# Realitzem el test de igualtat de variances. Utilitzem  Bartlett perquè sabem que les dues categories són normals. 

bartlett_test = stats.bartlett(Top_cat1['GA'],Top_cat2['GA'], Top_cat3['GA'])
bartlett_test
```

Assumim la igualtat de variances

```python
# ANOVA unidireccional entre grups (només hi ha una variable que s'utilitza per dividir el grup)
from scipy.stats import f_oneway

f_oneway(Top_cat1['GA'], Top_cat2['GA'], Top_cat3['GA'])
```

Rebutjem la hipòtesis nul·la i per tant els equips de les 3 categories tenen una mitjana diferent de gols en contra.
